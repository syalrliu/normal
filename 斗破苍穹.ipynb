### 词语出现频率搜索器

```python
a = 1
while a:
    a = input('输入:')
    rel=re.compile(a)
    rell=rel.findall(text)
    print(len(rell))
```

```python
words=jieba.lcut(text)
counts={}
for word in words:
   if len(word)==1:  #排除带个字符的分词效果
      continue
   else:
      counts[word]=counts.get(word,0)+1
items=list(counts.items())
items.sort(key=lambda x:x[1],reverse=True)
for i in range(15):
   word,count=items[i]
   print("{0:<10}{1:>5}".format(word,count))
```

```python
names=['萧炎','美杜莎','萧薰儿','小医仙','药老']
words_2=words.copy()
words_2=['萧薰儿' if i=='薰儿' else i for i in words_2]
words_2=['美杜莎' if i=='彩鳞' else i for i in words_2]
words_2=['药老' if i=='药尘' else i for i in words_2]
counts_2={}
for word in words_2:
    if len(word)==1:  #排除带个字符的分词效果
        continue
    else:
        if word in names:
            counts_2[word]=counts_2.get(word,0)+1      
```
